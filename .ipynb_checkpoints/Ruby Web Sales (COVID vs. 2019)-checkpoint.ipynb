```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplo as plt
import matplotlib
# %matplotlib inline
```

```python
true_web = pd.read_csv('Web_Sales_COVID.csv')

true_web = true_web.rename(columns={'SoldQty':'Qty',
                                    'ShipCost':'Ship Cost',
                                    'ProductCost':'Product Cost',
                                    'SubtotalAfterDiscount':'Subtotal'}) 

true_web.head()
```

```python
# Read in data for COVID web sales. Special encoding is needed

# covid_web = pd.read_csv('Ruby_websales_COVID.csv',encoding='latin-1')
```

```python
# covid_web
```

```python
true_web.dtypes
```

```python
total_covid_web_sales = true_web['Subtotal'].sum()
total_covid_web_sales
```

```python
avg_covid_price = (true_web['Subtotal']/true_web['Qty']).mean()
avg_covid_price
```

```python
covid_qty = true_web['Qty'].sum()
covid_qty
```

```python
# covid_web.sort_values(by='Qty',ascending=False).head()
```

```python
#make new column for the year 2020

true_web.insert(3,'Year',2020)
true_web
```

```python
# Read in data for pre-COVID web sales

pre_covid_web = pd.read_csv('PreCOVID_web_sales_2019.csv')
```

```python
pre_covid_web.head()
```

```python
pre_covid_web = pre_covid_web.rename(columns={'SoldQty':'Qty',
                                    'ShipCost':'Ship Cost',
                                    'ProductCost':'Product Cost',
                                    'SubtotalAfterDiscount':'Subtotal'}) 

# pre_covid_web
```

```python
#make new column for year 2019

pre_covid_web.insert(3,'Year',2019)
```

```python
total_precovid_web_sales = pre_covid_web['Subtotal'].sum()
total_precovid_web_sales
```

```python
avg_precovid_price = (pre_covid_web['Subtotal']/pre_covid_web['Qty']).mean()
avg_precovid_price
```

```python
precovid_qty = pre_covid_web['Qty'].sum()
precovid_qty
```

```python
join_df = [true_web,pre_covid_web]

web_sales = pd.concat(join_df)

```

```python
web_sales.head()
```

```python
# Calculate the %change of each of the categories

increase_sales = total_covid_web_sales-total_precovid_web_sales
pct_change_sales = (increase_sales/total_covid_web_sales).round(2)*100
print(pct_change_sales)  

increase_price = avg_covid_price-avg_precovid_price
pct_change_price = (increase_price/avg_covid_price)*100
print(pct_change_price)

increase_qty = covid_qty-precovid_qty
pct_change_qty = (increase_qty/covid_qty).round(2)*100
print(pct_change_qty)
```

```python
# Insert all into a new DataFrame

d = {'Pre-Covid':[total_precovid_web_sales,avg_precovid_price,precovid_qty],
     'COVID':[total_covid_web_sales,avg_covid_price,covid_qty],
     '% Change':[pct_change_sales,pct_change_price,pct_change_qty]}

Totals = pd.DataFrame(data=d,index=['Total Sales','Average Price','Quantity']).round(2)
```
