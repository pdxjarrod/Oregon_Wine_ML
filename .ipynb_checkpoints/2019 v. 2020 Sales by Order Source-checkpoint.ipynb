```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import cufflinks as cf
cf.go_offline
%matplotlib inline
```

```python
sales_source = pd.read_csv('order_sources_2019-2020.csv')
```

```python
sales_source.head()
```

```python
sales_source.drop(labels='Unnamed: 2',axis=1,inplace=True)
```

```python
sales_source
```

```python
grouped_source = sales_source.groupby(['Year']).sum()
```

```python
grouped_source
```

```python
# Display bar chart showing Net Revenue by Order Source and Year

# plt.use('Agg')

# plt.figure(figsize=(10,25))

sns_plot = sns.barplot(x='Order Source',y='Net Revenue',hue='Year',data=sales_source,palette='bright')
```

```python
sns_plot.figure.savefig('COVID_vs_2019_sales_sources.png')
```
