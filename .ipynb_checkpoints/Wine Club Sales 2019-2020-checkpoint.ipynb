```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import cufflinks as cf
cf.go_offline
%matplotlib inline
```

```python
wine_club = pd.read_csv('wine_club_sales_2019-2020.csv')
```

```python
wine_club.head()
```

```python
# wine_club.drop('Unnamed: 2',inplace=True,axis=1)

wine_club.rename(columns={'Group by Wine Club Name': 'Wine Club Name','Revenue': 'Gross Revenue'},inplace=True)
```

```python
wine_club
```

```python
wine_club['Qty'] = wine_club['Qty'].str.replace(',','')

wine_club['Qty'] = wine_club['Qty'].astype(float)
```

```python
wine_club['Average Revenue Per Unit'] = wine_club['Net Revenue']/wine_club['Qty'].astype(float)

wine_club['Discount'] = wine_club['Discount'] *(-1)

wine_club
```

```python
# Group the dataset by Year and sum the numerical columns

grouped_club = wine_club.groupby(['Year']).sum()
```

```python
grouped_club
```

```python
# Display the Net Revenue vs. Total Revenue

import matplotlib.pyplot as plt

f, axes = plt.subplots(2, figsize=(18, 18), sharex=False)

sns.set(rc={'figure.figsize':(15,8)})


sns_plot_1 = sns.barplot(x='Wine Club Name',y='Gross Revenue',hue='Year', data=wine_club,ax=axes[0])

sns_plot_2 = sns.barplot(x='Wine Club Name',y='Net Revenue',hue='Year', data=wine_club,ax=axes[1])

sns_plot_1.set_title('Gross vs. Net Revenue', fontsize=30)
```

```python
sns_plot_1.figure.savefig('Wine Club Gross vs Net Revenue 2019-2020')
```

```python
# Display Average Per Unit Revenue

avg_revenue_plot = sns.barplot(x='Wine Club Name',y='Average Revenue Per Unit',hue='Year', data=wine_club)

avg_revenue_plot.set_title('Average Revenue Per Unit by Club Level',fontsize=30)
```

```python
avg_revenue_plot.figure.savefig('Wine Club Average Revene Per Unit 2019-2020')
```
