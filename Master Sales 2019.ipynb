```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import datetime
%matplotlib inline
```

```python
# Read in 2019 sales dataset

master_19 = pd.read_csv('Data/Master_Sales_2019.csv')

# Read in 2020 sales dataset

master_20 = pd.read_csv('Data/master_sales_2020.csv')

# Read in 2018 sales dataset

master_18 = pd.read_csv('Data/master_sales_2018.csv')
master_18

```

```python
# insert year column into dataframse

master_19.insert(2,'Year',2019)

master_20.insert(2,'Year',2020)
```

```python
# Join 2019 & 2020 dataframes

new_master_df = pd.concat([master_19,master_20],join='inner')

new_master_df
```

```python
# Create bins to classify revenue ranges

bins = [ -1,0, 50, 250, 500, 1000, 5000,8000]
names = ['N/A','Very Low','Low', 'Medium', 'High', 'Very High','Ballers']

new_master_df["Buying Level"] = pd.cut(new_master_df['Revenue'],bins, labels=names)

new_master_df["Buying Level"].value_counts()
```

```python
# Display total revenue and discounts based on buying level

grouped_levels = new_master_df.groupby(['Buying Level']).sum().sort_values(by='Revenue',ascending=False).reset_index()

grouped_levels
```

```python
# Create a new df of totals by buying level
buying_level = grouped_levels['Buying Level'] 

buying_qty = grouped_levels['Qty']

avg_price = grouped_levels['Revenue']/grouped_levels['Qty']

avg_discount = grouped_levels['Discount']/grouped_levels['Qty']

buying_revenue = grouped_levels['Revenue']

buying_avg_revenue = grouped_levels['Revenue']/grouped_levels['Qty']


buyer_total_df = pd.DataFrame({'Buying Level':buying_level,
                              'Qty':buying_qty,
                              'Per Unit Revenue':avg_price,
                              'Total Revenue':buying_revenue,
                              'Average Discount':avg_discount
                              })

buyer_total_df.sort_values(by='Qty',ascending=False)
```

```python
# Create graph of 2019 - 2020 Buying Levels

buying_levels_plot = sns.countplot(x='Buying Level',hue='Year',data=new_master_df)

sns.set(rc={'figure.figsize':(15,8)})

buying_levels_plot.set_title('Total Transactions by Price Point',
                            fontsize=30)
```

```python
# Look at Customer Classes

new_master_df['Customer Class'].value_counts()
```

```python
# Display the number of transactions for each customer class

class_count = sns.countplot(x='Customer Class',hue='Year',
                            data=new_master_df,orient='v')

sns.set_context('talk')
sns.set(rc={'figure.figsize':(25,10)})
class_count.set_title('Transactions Per Customer Class',
                      fontsize=30)
```

```python
# Group df by Total Spent by Each Customer Class

class_total = new_master_df.groupby(['Customer Class','Year']).sum().reset_index()

class_total.head()
```

```python
# Create column of Customer Classes

cust_class = class_total['Customer Class'].sort_values()

# Create Column for total revenue, quantity, per unit revenue

cust_class_year = class_total['Year']

total_rev = class_total['Revenue']

total_qty = class_total['Qty']

per_unit_rev = class_total['Revenue']/class_total['Qty']

# new_master_df
```

```python
# Create new df showing revenues by customer class

class_revenues_df = pd.DataFrame({'Customer Class':cust_class,
                                  'Qty':total_qty,
                                  'Total Revenue':total_rev,
                                  'Per Unit Revenue':per_unit_rev,
                                  'Year':cust_class_year})

format_dict = {'Total Revenue':'${0:,.2f}','Per Unit Revenue':'${0:,.2f}','Qty':'{0:,.2f}'}

class_revenues_df.reset_index(drop=False,inplace=True)
```

```python
class_revenues_df.head().style.format(format_dict).hide_index()
```

```python
# Display Total Revenues by Each Class for 2019-2020

revenue_plot = sns.barplot(x='Customer Class',y='Total Revenue',
                           data=class_revenues_df,hue='Year')

revenue_plot.set_title('Total Revenue by Customer Class',
                       fontsize=30)
```

```python
# Display Per Unit Revenues by Each Class for 2019-2020

unit_revenue_plot = sns.barplot(x='Customer Class',y='Per Unit Revenue',data=class_revenues_df,hue='Year')

unit_revenue_plot.set_title('Per Unit Revenue by Customer Class',
                           fontsize=30)
```

```python
grouped_order_sources = new_master_df.groupby(['Year','Order Source']).sum().reset_index()

# grouped_order_sources
```

```python
# Create new df for Order Source revenue

order_source_years = grouped_order_sources['Year']

order_sources = grouped_order_sources['Order Source'].sort_values()

total_source_rev = grouped_order_sources['Revenue']


total_source_qty = grouped_order_sources['Qty']

per_unit_source_rev = grouped_order_sources['Revenue']/grouped_order_sources['Qty']

source_revenues_df = pd.DataFrame({'Order Source':order_sources,
                                  'Qty':total_source_qty,
                                  'Total Revenue':total_source_rev,
                                  'Per Unit Revenue':per_unit_source_rev,
                                  'Year':order_source_years})

format_dict = {'Total Revenue':'${0:,.2f}','Per Unit Revenue':'${0:,.2f}','Qty':'{0:,.2f}'}

source_revenues_df.sort_values(by='Total Revenue',ascending=False).reset_index(drop=True).style.format(format_dict).hide_index()

```

```python
# Display plot showing the count of transactions by order source
ax = sns.countplot(x='Order Source', data=new_master_df, hue='Year')

ax.set_title('Transactions by Order Source',
             fontsize=30)
```
